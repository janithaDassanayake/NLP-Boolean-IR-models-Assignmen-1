```python
# a). Write a program to identify the terms in the document
```

```python
#libraries

import os
import pandas as pd
```

```python

def document():
    wordsadded = {}

    os.chdir('D:/3rd 2nd SEM/subject work/IRWA/ASS1/Q1')
    cwd = os.getcwd()
    filelist = os.listdir(os.getcwd())

    for file in filelist:

        with open(file, 'r', encoding="utf8") as f:
            words = f.read().lower().split()

            for word in words:
                if word[-1] in ['.', ',', '!', ';', ':']:
                    word = word[-1]
                if word not in wordsadded.keys():
                    wordsadded[word] = [f.name]
                else:
                    if file not in wordsadded[word]:
                        wordsadded[word] += [f.name]

    return wordsadded, cwd, filelist

# a). Write a program to identify the terms in the document.

print('\n')
print('==========================')
print('a) Write a program to identify the terms in the document')
print('==========================')
print()
values = document()[0]
terms = []
for term in values.keys():
    terms.append(term)
print(terms)
```

```python
# b). Build a term document incidence matrix
```

```python
cwd = os.getcwd()
os.chdir('D:/3rd 2nd SEM/subject work/IRWA/ASS1/Q1')
file_List = os.listdir(os.getcwd())


def Matrix_one():
    dict_one = {}
    with open(file_List[0], 'r') as f:
        words = f.read().lower().split()
        unique_words = set(words)
        List_of_values = document()[0]
        for terms in sorted(List_of_values.keys()):
            if terms in unique_words:
                status = 1
            else:
                status = 0
            dict_one[terms] = status
    return dict_one


def Matrix_two():
    dict_two = {}
    with open(file_List[1], 'r') as f:
        words = f.read().lower().split()
        unique_words = set(words)

        List_of_values = document()[0]
        for terms in sorted(List_of_values.keys()):
            if terms in unique_words:
                status = 1
            else:
                status = 0
            dict_two[terms] = status
    return dict_two


def Matrix_three():
    dict_three = {}
    with open(file_List[2], 'r') as f:
        words = f.read().lower().split()
        unique_words = set(words)
        List_of_values = document()[0]
        for terms in sorted(List_of_values.keys()):
            if terms in unique_words:
                status = 1
            else:
                status = 0
            dict_three[terms] = status
    return dict_three


wordsadded = {'*    D1    *': Matrix_one(), '    D2    *': Matrix_two(), '   D3    *': Matrix_three()}
df = pd.DataFrame(wordsadded, columns=['*    D1    *', '    D2    *', '   D3    *'])
print('\n')
print('==========================')
print('b). Build a term document incidence matrix')
print('==========================')
print()
print(df)
```
