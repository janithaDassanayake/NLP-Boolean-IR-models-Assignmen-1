```python
# sentences tokenize

def jani_sentences_tokenization():
    cwd = os.getcwd()
    sentence = []
    os.chdir('C:\\Users\\94710\\AppData\\Roaming\\nltk_data\\corpora\\reuters\\training')
    fileList = os.listdir(os.getcwd())
    splitted_sentence = []
    for file in fileList:
        with open(file, 'r') as f:
            content = f.read()
            splitted_sentence += re.split("(?<=[.!?])\s+", content)
            sentence.append(splitted_sentence)
    return sentence
```

```python

"""
# Create word_tokenization  .
"""

def word_tokenization(word_tokens):
    symbols = ["!", "@", "#", "$", "%", "^", "&", "*", "(", ")", "~", "`", "-", "_", "=", "+", "[", "]", "{", "}",
               ";", ":", "'", '""', "?", "/", ">", "<", ".", ","]
    word_token = word_tokens.split()
    lists = []
    for word in word_token:
        for i in symbols:
            word = word.replace(i, '')
        lists.append(word)
    return lists
```

```python
# 2.	Build module for Porter Stemmer to do stemming.	
```

```python
"""
# Create potter_Stemmer alorithem  .
"""


def potter_Stemmer(lll):
    potter_Stemmered_word = []

    for tk_Word in lll:

        if tk_Word.endswith('sses'):
            inword = tk_Word.rfind('sses')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('sses', 'ss')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ies'):
            inword = tk_Word.rfind('ies')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ies', 'i')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ss'):
            inword = tk_Word.rfind('ss')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ss', 'ss')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('s'):
            inword = tk_Word.rfind('s')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('s', ' ')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ational'):
            inword = tk_Word.rfind('ational')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ational', 'ate')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('tional'):
            inword = tk_Word.rfind('tional')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('tional', 'tion')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ization'):
            inword = tk_Word.rfind('ization')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ization', 'ize')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ation'):
            inword = tk_Word.rfind('ation')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ation', 'ate')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ator'):
            inword = tk_Word.rfind('ator')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ator', 'ate')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('alism'):
            inword = tk_Word.rfind('alism')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('alism', 'al')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ement'):
            inword = tk_Word.rfind('ement')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ement', '')
                potter_Stemmered_word.append(tk)

        elif tk_Word.endswith('ent'):
            inword = tk_Word.rfind('ent')
            if inword == 1:
                potter_Stemmered_word.append(tk_Word)
            elif inword > 1:
                tk = tk_Word.replace('ent', '')
                potter_Stemmered_word.append(tk)
        else:
            potter_Stemmered_word.append(tk_Word)

    return potter_Stemmered_word
```

```python
# 3.	Build module to apply case folding i.e. convert words to lower case. 
```

```python

"""
# Create case_folding
"""


def case_folding(Stemmered_words):
    lower_List = []
    for Stemmered_word in Stemmered_words:
        result = ''
        for char in Stemmered_word:
            if char not in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
                result = result + char
            else:
                a = ord(char)
                b = a + 32
                result = result + chr(b)
        lower_List.append(result)
    return lower_List
```

```python
# 4.	Build inverted index and posting list as python dictionary.	
```

```python
#inverted index 
"""
# Create inverted index.
"""


def inverted_index():
    words_dict = {}
    cwd = os.getcwd()
    os.chdir('C:\\Users\\94710\\AppData\\Roaming\\nltk_data\\corpora\\reuters\\janitha')

    file_List = os.listdir(os.getcwd())
    for single_file in file_List:

        with open(single_file, 'r') as f:
            para = f.read()

            tokens_words = word_tokenization(para)

            pd = potter_Stemmer(tokens_words)

            lower = case_folding(pd)

            for single_word in lower:
                if single_word not in words_dict.keys():
                    words_dict[single_word] = [f.name]
                else:
                    if single_file not in words_dict[single_word]:
                        words_dict[single_word] += [f.name]

    return words_dict


print('\n')
print('==========================')
print('b). posting list as python dictionary')
print('==========================')
print()
for i in inverted_index().items():
    print(i)
print()
```
