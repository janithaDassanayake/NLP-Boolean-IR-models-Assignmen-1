```python
import numpy


def print_levenshtein_Distances(LD, string1Length, string2Length):
    for charIndex1 in range(string1Length + 1):
        for charIndex2 in range(string2Length + 1):
            print(int(LD[charIndex1][charIndex2]), end=" ")
        print()


def compute_levenshtein_Distance(string1, string2):
    LD = numpy.zeros((len(string1) + 1, len(string2) + 1))

    for charIndex1 in range(len(string1) + 1):
        LD[charIndex1][0] = charIndex1

    for charIndex2 in range(len(string2) + 1):
        LD[0][charIndex2] = charIndex2

    A1 = 0;  A2 = 0; A3 = 0;

    for charIndex1 in range(1, len(string1) + 1):
        for charIndex2 in range(1, len(string2) + 1):
            if string1[charIndex1 - 1] == string2[charIndex2 - 1]:
                LD[charIndex1][charIndex2] = LD[charIndex1 - 1][charIndex2 - 1]
            else:
                A1 = LD[charIndex1][charIndex2 - 1]
                A2 = LD[charIndex1 - 1][charIndex2]
                A3 = LD[charIndex1 - 1][charIndex2 - 1]

                if A1 <= A2 and A1 <= A3:
                    LD[charIndex1][charIndex2] = A1 + 1
                elif A2 <= A1 and A2 <= A3:
                    LD[charIndex1][charIndex2] = A2 + 1
                else:
                    LD[charIndex1][charIndex2] = A3 + 1

    print_levenshtein_Distances(LD, len(string1), len(string2))
    return LD[len(string1)][len(string2)]


def write_levenshtein_distance():
    word = compute_levenshtein_Distance(w1, w2)
    word_str = str(word)
    f = open("levenshtein_distance.txt", "a")
    f.write('\n\n')
    f.write('the distance of (' + w1 + ' & ' + w2 + ') = ' + word_str)
    f.close()


w1 = ''
print(' Type quit if you want to Exit')
print('=================================')
print()
w1 = input("Input the word 1 : ")

while w1 != 'quit':
    if w1 == 'quit':
        break
    else:
        w2 = input("Input the word 2 : ")
    if w2 == 'quit':
        break
    print()
    answer = compute_levenshtein_Distance(w1, w2)
    print()
    print('The levenshtein distance = ', answer)
    print()
    w1 = input("Input the word 1 : ")



```

```python
# Q3).
# 3 Implement the SOUNDEX algorithm.


def find_soundex_code(given_word: str):

    given_word = given_word.lower()
    letter_list = [character for character in given_word if character.isalpha()]
    

    # Step 1: -------------------------------------------------------------------------------
   

    if len(given_word) == 1:
        return given_word + "000"

    
    # a=0  /  e=0  /  i=0  /  o=0  /  u=0  /  y=0  /  h=0  /  w=0
    Replace_by_zero = ('a', 'e', 'i', 'o', 'u', 'y', 'h', 'w')

    first_letter = letter_list[0]
    letter_list = letter_list[1:]
    letter_list = [character for character in letter_list if character not in Replace_by_zero]

    if len(letter_list) == 0:
        return first_letter + "000"

    # Step 2:-------------------------------------------------------------------------------

    should_replace = {('b', 'f', 'p', 'v'): 1, ('c', 'g', 'j', 'k', 'q', 's', 'x', 'z'): 2,
                      ('d', 't'): 3, ('l',): 4, ('m', 'n'): 5, ('r',): 6}

    
    first_letter = [value if first_letter else first_letter for letter_group, value in should_replace.items()
                    if first_letter in letter_group]

    
    letter_list = [value if character else character
                   for character in letter_list
                   for letter_group, value in should_replace.items()
                   if character in letter_group]

    # Step 3:----------------------------------------------------------------------------------
    
    letter_list = [character for ind, character in enumerate(letter_list)
                   if (ind == len(letter_list) - 1 or (ind + 1 < len(letter_list) and character != letter_list[ind + 1]))]
    
    

    # Step 4:----------------------------------------------------------------------------------
    
    
    if first_letter == letter_list[0]:
        letter_list[0] = given_word[0]
    else:
        letter_list.insert(0, given_word[0])

        
    # Step 5:---------------------------------------------------------------------------------

    first_letter = letter_list[0]
    letter_list = letter_list[1:]

    letter_list = [character for character in letter_list if isinstance(character, int)][0:3]

    while len(letter_list) < 3:
        letter_list.append(0)

    letter_list.insert(0, first_letter)

    code = "".join([str(l) for l in letter_list])

    return code.upper()



def write_soundex():
    word = find_soundex_code(w1)
    word_str = str(word)
    f = open("soundex.txt", "a")
    f.write('\n\n')
    f.write('the soundex code for (' + w1 + ') = ' + word_str)
    f.close()

    
w1 = ''
print(' Type quit if you want to Exit')
print('=================================')
w1 = input("Input the word  : ")
print()
while w1 != 'quit':
    if w1 == 'quit':
        break
    print()
    answer = find_soundex_code(w1)
    write_soundex()
    print('The Soundex code = ', answer)
    print('------------------------')
    print()
    w1 = input("Input the word  : ")
```
